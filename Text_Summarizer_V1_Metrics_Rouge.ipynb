```python
from rouge import Rouge 
rouge = Rouge()
import swifter
from tqdm import tqdm
tqdm.pandas()

import dask.dataframe as dd
ddf_compare = dd.from_pandas(df_compare.fillna(""), npartitions=30)
from dask.diagnostics import ProgressBar
ProgressBar().register()
```

```python
def calc_rouge(final_summary,devopedia_summary):
    try:
        rouge_score=rouge.get_scores(final_summary,devopedia_summary)
        return rouge_score
    except:
        return "Error"
```

```python
%%time
# Generating Rouge Score
df_compare_processed=ddf_compare.map_partitions(lambda df: df.assign(rouge_scores=calc_rouge(df.final_summary, df.devopedia_summary))).compute()
```

```python
df_compare_processed
```

```python
df_compare_processed.to_csv("rouge_scores_4.csv",index=False)
```

```python
df_compare_processed.rouge_scores[0]
```

```python
# Extracting rouge score values from dictionary to Series
rouge_split=df_compare_processed['rouge_scores'].apply(pd.Series)
rouge_1=rouge_split['rouge-1'].apply(pd.Series).rename(columns={"f":"f1_r1",'r':"r_r1","p":"p_r1"})
rouge_2=rouge_split['rouge-2'].apply(pd.Series).rename(columns={"f":"f1_r2",'r':"r_r2","p":"p_r2"})
rouge_L=rouge_split['rouge-l'].apply(pd.Series).rename(columns={"f":"f1_rL",'r':"r_rL","p":"p_rL"})
```

```python
rouge_L.columns
```

```python
rouge_scores_all_df=pd.concat([df_compare_processed,rouge_1,rouge_2,rouge_L],axis=1).drop(["rouge_scores"],axis=1)
```

```python
rouge_scores_all_df.to_csv("rouge_scores_4.csv",index=False)
```
